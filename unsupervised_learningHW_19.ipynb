### a) **Classification:**

* Classification is a task of Machine Learning which assigns a label value to a specific class and then can identify a particular type to be of one kind or another. In Classifcation, **our target variable will be discrete**. It might be binary class like "spam" or "no spam", or multi class like "high","low","medium".

1) **KNN(KNearest Neighbors)classifier:**

* KNN algorithm works like majority voting based on nearest samples. **Choosing n_neighbors(no of neighbors)** is important. To avoid tie, give odd values.Sometimes we can choose k value as square root of(n), where n is total number of data points in the dataset.

* When new observation enters in the dataset, KNN search its nearest neighbor to determine which class the new observation will belongs to.

     **Notable Hyperparameters:**

     * n_neighbors

How does KNN works?

   * Finding Euclidean Distance between given element with the neighbors.
   * Get the least distance neighbour (closest neighbor)
   * Make prediction based on the closest neighbor
   
2) **Logistic Regression**

* Its a classification algorithm not a regression. Logistic regression is a statistical analysis method to predict a binary outcome, such as yes or no, based on prior observations of a data set.

* Since it predicts probability, the output values lies between 0 and 1.(probability always lies between 0 to 1)

     **Notable Hyperparameters are:**

   * C - Regularization parameter
   * penalty - Type of regularization(l1, l2, elasticnet, none}
   
   
3) **Naive Bayes Classifier**

* This classifier depends on Bayes theorem based on assumptions of independent variables.That means changing the value of one feature doesn't affect other feature.

Why is it called Naive?

* It is called Naive because of the assumption that 2 variables are independent when they may not be. In a real-world scenario, there is hardly any situation where the features are independent.

    **Notable Hyperparameters GaussianNB are:**
    * priors
    * var_smoothing
    
4) **Support Vector Machines** 

* Main Objective of Support Vector Machine is **find a hyperplane for an N Dimensional features that classify the data points.** Hyper planes are decision boundaries that help classify the data points.

* Datas which on or near to the boundary (hyper plane) are called support vectors. Hence the name become Support Vector Machines. Hyper plane which contains maximum distance between support vectors is the best hyperplane we can select.

* SVM has a special property **it uses only the support vectors and the rest of the data points can be removed without affecting fitting the model.** All Other classifiers uses all data points while fitting the model.

     **Notable Hyperparameters are:**
    * C - Regularization parameter
    * kernel - which is linear or rbf
    * gamma - only applies to rbf kernel
    
5) **Decision Tree classifier**

* Decision tress is a tree shaped diagrams. It asks a question and then classify the sample based on the answer.

* To find solutions decision tree makes hierarchical decision based on the data. Hierarchical means the model is defined by  a series of questions to any observations.Eg.,"if this occurs then if this occurs".

How this works?
    
   * **It will start at the top, works it way down and down until it get to a point where it can't go further**. Thats how it classifies the samples.
   
        **Notable Hyperparameters are:**
       * criterion - gini or entropy

6) **Random Forest classifier**

* Random Forest is a **bagging(bootstrap aggregating)** technique. Base estimator is decision tree. RF combines various decision trees to produce more generalized model. We can reduce high variance using bagging.

### Unsupervised Learning

* In Unsupervised learning **we donot supervise model**. We let the model to work on its own to discover informations since we are giving unlabeled datas.
* Machine must be able to **classify the data without any prior informations about data**. Discovering hidden patterns in data.

Some Unsupervised Learning applications are: 

   * Fraud Detection
   * Malware Detection
   * Customer Segmentation
   
Types:

   1) Clustering
   
        a) KMeans Clustering
        b) Hierarchical Clustering
        
   2) Dimensionality Reduction
   
        a) PCA
        b) SVD
        c) LDA
        d) Isomap Embedding
        
### 1) **Clustering**

* The method of dividing the objects into clusters(groups) based on its similarities between them and dissimilar to the objects which are belongs to another cluster.

* **Inertia measures clustering quality** How spread out the samples within each cluster can be measured by the "inertia". Intuitively, inertia measures how far samples are from their centroids.

* A good clustering has tight clusters, that is samples in each cluster are bunched together, not spread out. So lower values of the inertia are better.

a) **KMeans Clustering**
    
* K means is a simple unsupervised learning technique. This algorithm identifies k number of centroids, and then allocates every data point to the nearest cluster, while keeping the centroids as small as possible.

* means- refers to averaging of the data; that is, finding the centroid. Before we apply k-means, **we have to find optimal value of k. Elbow method with inertia, distortion are used to find optimal no of clusters**. A good rule of thumb is to choose an elbow in the inertia plot, a point where the inertia begins to decrease more slowly.
    
    **Notable Hyperparameter**
    * n_clusters             
        
b) **Hierarchical Clustering**
    
* Hierarchical clustering is separating data into different groups based on its similarities.
Two types:
    * Agglomerative Clustering (Bottom up approach)
    * Divisive Clustering (Top to Bottom approach)
   
How does hierarchical clustering works?

   * Each data point is a cluster of its own in Agglomerative Clustering.
   * We find the least distance between two data points or cluster.
   * The nearest datapoints are clustered together. This is represented in a tree structure called **dendogram**
   * We terminate when we are left with one cluster. 
 
### 2) **Dimensionality Reduction Techniques**

* Dimensionality reduction is used to reduce dimensions of our inputs from a dataset.

* It won't just pick columns from the dataset like feature selection, **it will reduce the dimensions by creating new columns using datas from the dataset**.

Why it is used?

* The more features we use in our model, will leads to **overfitting problem**. Dimensionality reduction helps us to reduce overfitting by keeping the more important features in the feature set and reducing the number of features that do not help to decide the output. 
    
a) **PCA(Principal Component Anaysis)**

* PCA will give more preference to high variance features and then it will go to low variance. For instance, if n_components = 3, we are asking PCA to extract the most 3 important components. Principal Component 1(PC1) contains the most variance features, PC2 contains second most highest variance features and so on.  

* We need to find the exact n_components value. Using **explained_variance_ parameter we can find the best n_components**. 

b) **LDA(Linear Discriminant Analysis)**

* LDA handles multiple class problem very efficiently. It provides an informative low-dimensional view on the data, which is both useful for visualization and feature engineering.

* But if the classes are non-linearly separable, It can not find a lower-dimensional space to project. 

* The number of dimensions is limited to 1 and C-1, where C is the number of classes. If Our dataset is a binary classification problem, then the number of dimensions is 1. The “n_components” argument can be set to configure the number of desired dimensions in the output of the transform. So for binary case n_components = 1

c) **Isomap Embedding**

* Isomap Embedding is a non linear dimnensionality reduction. It preserves non linear relationships in the dataset.

* Isomap techniques uses Knearest neighbors to compute shortest path and construct the neighborhood graph. It uses multi dimensional scaling to compute lower dimensional embedding.

* The “n_components” argument can be set to configure the number of desired dimensions in the output of the transform.

#### Adding the column - Amount_spent

```python
# we need to know how much total amout each customer spent. I am creating a feature called total cost by multiplying
# unit price and quantity

ecommerce_new_df['Amount_spent'] = ecommerce_new_df['Quantity'] * ecommerce_new_df['UnitPrice']
ecommerce_new_df['Amount_spent']
```

#### Extracting month and year from Invoice date 

```python
ecommerce_new_df['InvoiceDate'] = pd.to_datetime(ecommerce_new_df['InvoiceDate'])
ecommerce_new_df['Year'] = pd.DatetimeIndex(ecommerce_new_df['InvoiceDate']).year
ecommerce_new_df['Month'] = pd.DatetimeIndex(ecommerce_new_df['InvoiceDate']).month
ecommerce_new_df.head()
```

* Since this is customer segmentation, customers are grouped according to their purchases. Keep the relevant information for each customer:
    * total amount of purchases
    * date(Month, year)
    * how many different products the customer bought
    * total money spent

#### i) Total amount of purchases
* To find total no of purchases, we have to count invoice no for each customer id with grouping by customer id

```python
purchases = ecommerce_new_df[['CustomerID','InvoiceNo']].groupby(['CustomerID']).count()
purchases.rename(columns={'InvoiceNo':'total_purchases'}, inplace=True)
purchases
```

#### ii) date(Month, year)
* I have already extracted month and year from invoice date.

```python
start_month = ecommerce_new_df[['CustomerID','Month']].groupby(['CustomerID']).min()
start_month.rename(columns ={'Month': 'Start_Month'},inplace =True)
purchases = purchases.merge(start_month, how = 'inner', on = 'CustomerID')
```

```python
last_month = ecommerce_new_df[['CustomerID','Month']].groupby(['CustomerID']).max()
last_month.rename(columns = {'Month': 'End_Month'},inplace = True)
purchases = purchases.merge(last_month, how = 'inner', on ='CustomerID')
```

#### iii) How many different products the customer bought?

* To find the product count, I am subsetting customer id and stock code and then grouped by customer id.

```python
prod = ecommerce_new_df[['CustomerID','StockCode']].groupby(['CustomerID']).nunique()
prod.rename(columns = {'StockCode': 'Products'}, inplace = True)
purchases = purchases.merge(prod, how='inner',on='CustomerID')
```

 #### iv) total money spent

* Aggregating amount spent column, to find how much each customer spent with group by.


```python
money = ecommerce_new_df[['CustomerID','Amount_spent']].groupby(['CustomerID']).sum()
money.rename(columns = {'Amount_spent':'total money spent'},inplace = True)
purchases = purchases.merge(money, how = 'inner', on = "CustomerID")
```

```python
country = ecommerce_new_df[['Country','CustomerID']].groupby(['CustomerID']).min()
purchases = purchases.merge(country,how = 'inner', on = 'CustomerID')
purchases.head()
```

#### Standardizing

* Using scikit Normalizer I am going to change my datas to look normally distributed.

```python
scaled = purchases.drop('Country',axis =1)
from sklearn.preprocessing import Normalizer
sc = Normalizer()
scaled = sc.fit_transform(scaled)
scaled_df = pd.DataFrame(scaled, columns = ['total_purchases','Start_Month','End_Month','Products','total money spent'])
scaled_df
```

#### Hierarchical Clustering

Two types:

#### How does hierarchical clustering works?
   * Each data point is a cluster of its own in Agglomerative Clustering.
   * We find the least distance between two data points or cluster.
   * The nearest datapoints are clustered together. This is represented in a tree structure called **dendogram**

* So now, we can apply k-means over this information and see how the customers are grouped using clusters.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

#### Which approach did you use to find the ideal number of clusters and why?¶
 
* **Inertia measures clustering quality** How spread out the samples within each cluster can be measured by the "inertia". Intuitively, inertia measures how far samples are from their centroids.

* A good clustering has tight clusters, that is samples in each cluster are bunched together, not spread out. So lower values of the inertia are better.

* The inertia of a kmeans model is measured automatically after we fit our model with inertia attribute. In fact, kmeans aims to place the clusters in a way that minimizes the inertia. 

* **Before we apply k-means, we have to find optimal value of k**. I am choosing **Elbow method with inertia** to find optimal no of clusters. A good rule of thumb is to choose an elbow in the inertia plot, a point where the inertia begins to decrease more slowly. 

```python
inertias_list = []
for k in range(1,10):
    # instantiating KMeans
    k_cluster = KMeans(n_clusters = k)
    # fitting the model
    model = k_cluster.fit(scaled_df)
    inertia = model.inertia_
    inertias_list.append(inertia)
    
from kneed import KneeLocator
kn = KneeLocator(range(1,10),inertias_list, curve='convex', direction='decreasing')
print("Optimal No. of Clusters: ",kn.knee)

# plotting inertias
plt.plot(range(1,10),inertias_list, '-X')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(range(1,10))
plt.title('Elbow Method with Inertia')
plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dashed')
plt.show()
```

```python
# Now that we got our optimal n_clusters value, we can start clustering
cluster = KMeans(n_clusters = 3)
# fit the model
model1 = cluster.fit(scaled_df)
# creating labels for our input data
label = model1.predict(scaled_df)
# calculating cluster centers from model
cluster_centroids = model1.cluster_centers_
```

```python
plt.scatter(scaled_df['Start_Month'],scaled_df['total money spent'],c = label)
plt.title("Clusters based on customers total money spent and their shopping start month");
```

```python
plt.scatter(scaled_df['total_purchases'],scaled_df['End_Month'],c = label)
plt.title("Clusters based on customers total_purchases and their shopping end month");
```
